```python
# Para cuando es verdadero (SÍ))
# Calculando si el número es positivo if else
# Calculando si el número es negativo if else
# Daros de entrada (NÚM) Valor para determinar su signo, (R) Resultado del signo del valor
# Mostrar el numero es
NÚM = int(input("Ingrese un número: "))
if NÚM >= 0:
    R = "POSITIVO"
else:
    R = "NEGATIVO"
print(f"El número {NÚM} es {R}")
print("Programa realizado por Pedro Martinez NL = 34")
```

```python
# Para cuando es Verdadero (NO)
# Calculando si el numero es Positivo if else
# Calculando si el numero es Negativo if else
# Daros de entrada (NÚM) Valor para determinar su signo, (R) Resultado del signo del valor
# Mostrar el numero es
NÚM = int(input("Ingrese un número: "))
if NÚM >= 0:
    R = "POSITIVO"
else:
    R = "NEGATIVO"
print(f"El número {NÚM} es {R}")
print("Programa realizado por Pedro Martinez NL = 34")
```
